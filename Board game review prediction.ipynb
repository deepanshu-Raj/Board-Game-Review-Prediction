```python
#for plotting data
import matplotlib.pyplot as plt
import seaborn as sns

#splitting data for cross-validation
from sklearn.model_selection import train_test_split
```

```python
#load the data
games = pd.read_csv("games.csv")
```

```python
print(games.columns)
print(games.shape)
```

```python
#making histgram of all the ratings in the average_rating column
plt.hist(games["average_rating"])
plt.show()
```

```python
#if user_rating for any game = 0,no need to consider for that game.
#it simply means that no one has played that game.
games = games[games['users_rated']>0]

#removes any row with any missing data
games = games.dropna(axis = 0)
```

```python
plt.hist(games['average_rating'])
plt.show()
```

```python
print(games.columns)
```

```python
#corellation matrix
cormat = games.corr()
fig = plt.figure(figsize = (12,9))
sns.heatmap(cormat,vmax = 0.8 ,square = True)
plt.show()
```

```python
#get all the columns from the data frame
columns = games.columns.tolist()

#filter the columns to remove data, we do not want
columns = [c for c in columns if c not in ['bayes_average_rating','average_rating','type','name','id']]

#the values that we'll be predicting on.
target = 'average_rating'
```

```python
plt.figure(figsize = (8,6))
sns.scatterplot(x = columns[6],y = target,data = games)
plt.show()
```

# Observations :
### 1) max players are in the agegroup : (0-20) yrs old.
### 2) why(how) can a 0 year old ,play video game?
### CLAIM : reviews with minage = 0,can be treated as false!!

```python
#this data includes reviews for game which are qiute old.
print("{} : {}".format("total",games.shape[0]))
print("{} : {}".format("games plublished after 1750",games[games['yearpublished']>1750].shape[0]))

#plot between minage and average_rating.
plt.figure(figsize = (8,6))
#ci - confidence interval = "sd" = standard deviation
#the colouring is the error band colouring
sns.lineplot(x = columns[0],y = target,data = games[games['yearpublished']>1750],c = 'b')
plt.show()
```

# Conclusion:
### We can dump 3622 data-points on the point that reviews on the games that were published before year 1750, won't have much impact on the predictions that we are going to make today.

```python
train = games.sample(frac = 0.8,random_state = 1)

test = games.loc[~games.index.isin(train.index)]

print(train.shape)
print(test.shape)
```

```python
#used a linear model.
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

LR = LinearRegression()

#fit the model to the training data.
LR.fit(train[columns],train[target])
```

```python
#generate predictions for the test data
predictions = LR.predict(test[columns])

#compute error between test prediction and actual value
print("{} : {}".format("LR",mean_squared_error(predictions,test[target])))
```

```python
#import the random forest model
from sklearn.ensemble import RandomForestRegressor

#initialize the model
RFR = RandomForestRegressor(n_estimators = 100,min_samples_leaf = 10,random_state = 1)

#fit the data
RFR.fit(train[columns],train[target])
```

```python
predictions_2 = RFR.predict(test[columns])

print("{} : {}".format("RFR",mean_squared_error(predictions_2,test[target])))
```

### Conclusion : Non-linear model getting a better result than a linear model.
